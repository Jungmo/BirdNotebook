```python
import numpy as np
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt

n_groups = 4
bird = (30, 72, 93, 94.8)
egg = (32, 75, 86, 88)
empty = (99, 98, 96, 93)
child = (39, 67, 77, 76)

k8 = (30, 32, 99, 39)
k16 = (32, 75, 86, 88)
k32 = (93, 86, 96, 77)
raw = (94.8, 88, 93, 76)

fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8

rects1 = plt.bar(index, k8, bar_width,
                 alpha=opacity,
                 color='b',
                 label='k = 8')

rects2 = plt.bar(index + bar_width, k16, bar_width,
                 alpha=opacity,
                 color='g',
                 label='k = 16')

rects3 = plt.bar(index + bar_width*2, k32, bar_width,
                 alpha=opacity,
                 color='r',
                 label='k = 32')

rects4 = plt.bar(index + bar_width*3, raw, bar_width,
                 alpha=opacity,
                 color='k',
                 label='k = 256')

y = ("bird", "egg", "empty", "child")
y_pos = np.arange(len(y))


width = 0.35

plt.xticks(y_pos+bar_width*1.5, y)
plt.ylabel("Accuracy")
plt.title("Accuracy After Color Quantization")
plt.legend()
plt.show()
```
